# Class2: Data structure of R language

## data structure:
### 1. element (元素):
定义：数值型、字符串、逻辑型等都可称为元素。

### 2. vector (向量):

定义: 由 element 构成的 1-D 数据结构。

特点: 有序性、单个向量数据类型相同 (存储单一类型的数据)。

创建方法: `c()`

举例: `x = c(1,2,3,4)`

获取元素：`x[1:2]`, `x[c(1,3)]`

### 3. matrix (矩阵):()
创建方法：`matrix()`

举例： `x = matrix(c(3:14), nrow=4))`

矩阵转置：`t()`

获取元素:
``` R
x[2,3]

```
### 4. dataframe (数据框):

创建方法：`data.frame()`

举例：
``` R
x = data.frame(
    a = c("x","y","z"),
    b = c(1,2,3))

```

获取元素：
``` R
x[2,1]
x[2,c(1,2)]
x[:,c(4,3,2)]
```

修改列名：`colnames()`

修改行名：`rownames()`
